#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
#verificando a existência de dados ausentes
yf['Adj Close'].isna().sum()
````

````python
mm = yf['Adj Close']
mm.head()
````

````python
#obtendo as médias de Adj close
mm = yf['Adj Close'].rolling(90, min_periods=1).mean()
mm
````

````python
#calculando desvio padrão
dp = yf['Adj Close'].rolling(90, min_periods=1).std()
dp
````

````python
#pedindo o gráfico de média movel
mm_vertical = mm.stack().reset_index()
mm_vertical.columns = ['data', 'ativo', 'média móvel']

fig = px.line(mm_vertical, x="data", y="média móvel", color='ativo', template='plotly_white')
fig.show()
````

````python
#media móvel do desvio padrão
dp_vertical = dp.stack().reset_index()
dp_vertical.columns = ['data', 'ativo', 'desvio padrão']

fig = px.line(dp_vertical, x="data", y="desvio padrão", color='ativo', template='plotly_white')
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
#selecionando as variáveis
lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['PETR4.SA']*len(lista_indicadores)
````

````python
colunas = list(zip(lista_indicadores, ativo))
````

````python
PETR4_SA = yf.loc['10/04/2023':, colunas]
PETR4_SA.columns = lista_indicadores
````

````python
PETR4_SA.head()
````

````python
graph = {
    'x': PETR4_SA.index,
    'open': PETR4_SA.Open,
    'close': PETR4_SA.Close,
    'high': PETR4_SA.High,
    'low': PETR4_SA.Low,
    'type': 'candlestick',
    'name': 'PETR4_SA',
    'showlegend': True
}
````

````python
#gráfico de velas
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="PETR4_SA",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
